```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
#Functions:
plt.plot(x,y,'g-')
plt.xlabel('X - Label')
plt.ylabel('Y - Label')
plt.title('Fucntional')
plt.show()
```

```python
plt.subplot(3,1,1)
plt.plot(x,y,'b-') # where x = np.linspace(0,5,11)

plt.subplot(3,2,1)
plt.plot(x,y,'r-')# where y = x ** 2

plt.subplot(3,1,3)
plt.plot(x,y,'g-')# where y = x ** 2

plt.show()
```

```python
# object oriented
```

```python
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8]) #([left,bottom,width,height])
axes1 = fig.add_axes([0.23,0.55,0.4,0.2])

axes.plot(x,y,'r')
axes.set_title('A -Plot')
axes1.plot(y,x)
axes1.set_title('B-Plot')
```

```python
fig,axes = plt.subplots(nrows = 3,ncols = 2)

axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
# Figure Size
```

```python
fig,axes = plt.subplots(nrows = 2,ncols = 1,figsize = (8,3),dpi = 200)
axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x ** 2,label = 'X sq')
ax.plot(y,x*3,label = '3 times Y')

ax.legend(loc = 0)
```

```python

fig.savefig('my_picture.png',dpi = 200)
```
